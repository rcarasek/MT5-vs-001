```python
# CONFIGURANDO YAHOO FINANCE
!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
ativos = "MOSI3.SA GGBR4.SA"
# ativos = "IRBR3.SA BBSE3.SA"
dados_ativos = yf.download(tickers=ativos, period='1y')['Adj Close']
```

```python
retorno_ativos = dados_ativos.pct_change()
retorno_ativos_acumulado = (1 + retorno_ativos).cumprod()   # numpy.cumprod() function is used when we want to compute the cumulative product of array elements over a given axis.
retorno_ativos_acumulado.iloc[0] = 1
# retorno_ativos_acumulado.head()
```

```python
retorno_ativos_acumulado.plot(figsize=(18,8))
```
